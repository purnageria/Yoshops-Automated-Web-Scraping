```python
url = "https://yoshops.com/"
wanted_l = ["https://yoshops.com/t/toys"]

main_url_list = scraper.build(url, wanted_list=wanted_l)
```

```python
# To get names of categories :
def get_categories(list_url):
    cat = [ ]
    
    for i in list_url:
        word_list_i = i.split("/")
        word_i = word_list_i[-1:]
        cat.append(word_i)
    return cat



# To make a list of categories
def list_cat(cat):
    cate = []
    for i in cat:
        cat_i = i[0]
        cate.append(cat_i)
    return cate
```

```python
main_cat = get_categories(main_url_list)

main_categories = list_cat(main_cat)
```

```python
print("Main Categories :")
for i in main_categories:
     print(main_categories.index(i)+1, ". ", i.upper()) 
    
```

```python
def Get_sub_url():
    Enter_2 = int(input("Enter Required category Number :"))
    if Enter_2 in range(0,len(main_categories)+1):
        j = Enter_2-1
        category_url = main_url_list[j]
    else : 
        return Get_sub_url()
    
    from urllib.request import Request, urlopen
    from bs4 import BeautifulSoup as soup
    req = Request(category_url , headers={'User-Agent': 'Mozilla/5.0'})

    webpage = urlopen(req).read()
    page_soup = soup(webpage, "html.parser")
    
    urls = []
    for link in page_soup.find('div',class_='col-sm-3').find("li", class_ = 'active').find_all('a', href=True):
        urls.append(url[:-1]+link['href'])
     
    return urls
    
    
```

```python
urls = Get_sub_url()
```

```python
req1 = Request(urls[0] , headers={'User-Agent': 'Mozilla/5.0'})

webpage1 = urlopen(req1).read()
page_soup1 = soup(webpage1, "html.parser")
```

```python
z = []
for i in page_soup1.find_all("a", class_= "product-link", href = True):
    z.append("https://yoshops.com"+i["href"])
```

```python
def Product_Details():
    z = []
    for i in page_soup1.find_all("a", class_= "product-link", href = True):
        z.append("https://yoshops.com"+i["href"])
        
    name = []
    detail = []
    
    for i in z:
        from urllib.request import Request, urlopen
        from bs4 import BeautifulSoup as soup
        url = i
        req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})

        webpage = urlopen(req).read()
        page_soup = soup(webpage, "html.parser") 
    
        n = page_soup.find("div", class_ = "col-sm-5").find("h1").get_text()
        d = page_soup.find("div", class_ = "col-sm-5").find("div", class_="single-product-description")
        d = d.get_text().replace("\n", "")
        d = d.replace("\xa0", "")
        d = d.replace("\r", "")
        d = d.split("Product Description")
        name.append(n)
        detail.append(d[0])
        
    df = pd.DataFrame(list(zip(z, name, detail)), 
                 columns=["Web Url", "Product Name", "Product Detail"])
    df.to_excel("Product Details.xlsx", index = False)
    return df
```
